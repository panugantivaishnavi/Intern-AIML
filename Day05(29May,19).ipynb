### Problem Statement: Print number of matches won and loss(as two columns). Consider as win when a team wins Toss  and Match . print for each and every team

```python
import pandas as pd
import numpy as np
def winloss(filename):
    df=pd.read_csv(filename)
    team=[]
    uniqteam=[]
    for row in df.values:
        team.append(row[9])
    uniqteam=np.unique(team)    
    winlist=[]
    losslist=[]
    for i in uniqteam:
        wc=0
        lc=0
        for row in df.values:
            if(i==row[9]) and (i==row[10]) :
                wc=wc+1
            elif(row[2]==i or row[3]==i):
                lc=lc+1
        winlist.append(wc)
        losslist.append(lc)
    data={'Teams':uniqteam,"win":winlist,'loss':losslist}      
    df1=pd.DataFrame(data)
    return df1
winloss('data/Match.csv')
```
