## Importing all the required libraries

## Using Interpolate method for handling missing values in pandas

```python
purchase=purchase.interpolate() 
purchase
```

## Creating dummy sets using pd.get_dummies(dataframe.colnname) for a particular column
### Dummies are to be created for categorical data ie coln that has texual data into numerical data so that it woul be easy to scale all the coln of dataframe to one scale for creating the model in future .By dong this ,the efficiency of the model increases.

```python
dummy_countryset = pd.get_dummies(purchase.Country)
dummy_countryset
```

## Merging the created dummyset 'ie dummy_countryset' with dataframe"purchase" using concat method and axis specifier=1 ie merge coln wise.

```python
merge_country=pd.concat([purchase,dummy_countryset],axis=1)
merge_country
```

## Now remove a coln among the 3 and the country coln using dataframe.drop(['parent colnnme','child colnnme'],axis=1)

```python
fi=merge_country.drop(['Country','France'],axis=1)
fi
```

## Now create dummies for the second categorical coln using pd,get_dummies(dataframe.colnnme)

```python
dummy_purchasedset = pd.get_dummies(purchase.Purchased)
dummy_purchasedset
```

## Now merge the dummies with existing dataframe using concat.

```python
merge_purchase=pd.concat([fi,dummy_purchasedset],axis=1)
merge_purchase
```

## Now drop ie remove the parent coln and one child coln.It is industry rule that we hav to delete one child coln in created dummies and the parent coln also to increase efficiency.

```python
fii=merge_purchase.drop(['Purchased','Yes'],axis=1)
fii
```

## Splitting the data
