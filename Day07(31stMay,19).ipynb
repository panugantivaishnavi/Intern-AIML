## Problem 1:
### Problem Statement:Draw a pie plot based on number of wins for each team

```python
import pandas as pd
import numpy as np
class prob1():
    def win(self,filename):
        df=pd.read_csv(filename)
        self.mtchwin,self.matchwin_uniq,self.winlist,self.dicts=[],[],[],[]
        for row in df.values:
             self.mtchwin.append(row[10])
        #mtchwin=df.match_winner
        self.matchwin_uniq=np.unique(self.mtchwin)
        for i in range(0,len(self.matchwin_uniq)):
            win=0
            for j in range(0,len(self.mtchwin)):
                if self.matchwin_uniq[i]==self.mtchwin[j]:
                    win+=1
            self.winlist.append(win)
        i=self.plot()
    def plot(self):
        import matplotlib.pyplot as plt
        labels=self.matchwin_uniq
        sizes=self.winlist
        plt.pie(sizes,labels=labels,autopct='%1.1f%%',shadow=True,startangle=0)
        plt.title('Pie plot based on number of wins for each team',fontsize=20)
        plt.axis('equal')
        plt.show()
        
p=prob1()
p.win('data/Match.csv')
```

## Problem 2:
### Problem Statement:Draw a bar graph for number of matches won (i,e.consider win if the team wins both the toss and match)  for the team “Mumbai Indians” for all years

```python
import pandas as pd
import numpy as np
class prob2():
    def matchwonMI(self,filename):
        df=pd.read_csv(filename)
        self.yr,self.yruniq,self.matchwin,self.winlist=[],[],[],[]
        self.matchwin=df.match_winner
        self.toss=df.Toss_Winner
        self.yr=df.Season_Year
        self.yruniq=np.unique(self.yr)
        for i in range(0,len(self.yruniq)):
            self.win=0
            for j in range(0,len(self.yr)):
                if self.yr[j]==self.yruniq[i] and self.matchwin[j]=='Mumbai Indians' and self.toss[j]=='Mumbai Indians':
                    self.win+=1
            self.winlist.append(self.win)
        self.dicts=dict(zip(self.yruniq,self.winlist))
        #print(self.dicts)
        i=self.bars()
    def bars(self):
        import matplotlib.pyplot as plt
        index=self.yruniq
        plt.bar(self.yruniq,self.winlist)
        plt.xlabel('Years',fontsize=6)
        plt.ylabel('Number of matches won',fontsize=6)
        plt.xticks(index,fontsize=6,rotation=30)
        plt.title('Number of matches won by “Mumbai Indians” for all years',fontsize=16)
        plt.show()
p2=prob2()
p2.matchwonMI('data/Match.csv')
            
```
